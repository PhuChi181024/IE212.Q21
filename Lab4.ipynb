#### 3. Phân tích số lượng đơn hàng theo quốc gia, sắp xếp theo thứ tự giảm dần.

```python
orders_by_countries = orders_df.join(customer_df, "Customer_Trx_ID").select("Order_ID", "Customer_Country")
orders_by_countries_count = orders_by_countries.groupBy("Customer_Country").count()
for row in orders_by_countries_count.head(10):
    print(row)
```

#### 4. Phân tích số lượng đơn hàng nhóm theo năm, tháng đặt hàng (Hiển thị theo năm tăng dần, tháng giảm dần)

```python
orders_time = orders_df.select("Order_ID",
                               year(to_timestamp("Order_Purchase_Timestamp")).name("Year"), 
                               month(to_timestamp("Order_Purchase_Timestamp")).name("Month")
)
orders_time.take(3)
```

```python
orders_year_month = orders_time.groupBy("Year", "Month").count().orderBy(["Year","Month"], ascending = [True, False])
orders_year_month.take(5)
```

#### 5. Thống kê điểm đánh giá trung bình, số lượng đánh giá theo từng mức (ví dụ: 1 đến 5).
##### (Lưu ý: Cần xử lý các giá trị ngoại lệ và NULL trong cột Review_Score)

##### Xử lý các giá trị ngoại lệ và NULL trong cột Review_Score

```python
reviews_clean = reviews_df.select(expr("try_cast(Review_Score as int)").name("Review_Score"))
reviews_clean = reviews_clean.filter(reviews_clean["Review_Score"].isNotNull()).filter((reviews_clean["Review_Score"] >= 1) & (reviews_clean["Review_Score"] <= 5))
```

##### Thống kê điểm đánh giá trung bình và số lượng từng mức đánh giá

```python
avg_rating_count = reviews_clean.select(avg("Review_Score").alias("Average_Review_Score"))
avg_rating_count.show()

rating_count = reviews_clean.groupBy("Review_Score").agg(count("Review_Score").alias("Total_Reviews"))
rating_count = rating_count.orderBy("Review_Score")
rating_count.show()
```

#### 10.Xếp hạng các seller dựa trên tổng doanh thu và số lượng đơn hàng bán được

```python
# Nhóm theo đơn hàng theo Seller_ID
rank_sellers = items_df.groupBy("Seller_ID")
rank_sellers = rank_sellers.agg(sum("Price").alias("Total_Revenue"), count_distinct("Order_ID").alias("Total_Orders"))
rank_sellers = rank_sellers.orderBy(["Total_Revenue", "Total_Orders"], ascending=[False, False])
rank_sellers.show()
```
