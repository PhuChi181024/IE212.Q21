```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

#### Tạo map (MovieID -> List of Genres).

```python
movies = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/movies.txt")
movies_genres_rdd = movies.map(lambda x: x.split(",")).map(lambda x: ((x[0]), x[2].split("|")))
movies_genres_rdd.collect()
```

```python
ratings1 = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/ratings_1.txt")
ratings2 = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/ratings_2.txt")
ratings = ratings1.union(ratings2)
```

#### Map từ MovieID -> Rating -> (Genre, Rating).

```python
ratings_rdd = ratings.map(lambda x: x.split(",")).map(lambda x: ((x[1]), float(x[2])))
ratings_rdd.collect()
```

```python
movie_rating = movies_genres_rdd.join(ratings_rdd)
movie_rating.collect()
```

```python
genre_rating = movie_rating.flatMap(lambda x: [(genre, x[1][1]) for genre in x[1][0]])
for row in genre_rating.collect():
    print(row)
```

```python
# Tính tổng điểm đánh giá của mỗi genre
sum_rdd = genre_rating.reduceByKey(lambda a, b: a + b)
sum_rdd.collect()
```

```python
# Tính tổng số lượt đánh giá của mỗi genre
count_rdd = genre_rating.mapValues(lambda x: 1).reduceByKey(lambda a, b: a + b)
count_rdd.collect()
```

```python
average_rdd = sum_rdd.join(count_rdd).mapValues(lambda x: x[0] / x[1]) 
average_rdd = average_rdd.sortByKey()
for row in average_rdd.collect():
    print(row)
```
