```python
import findspark 
findspark.init()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

#### Đọc file movies.txt và tạo một map (MovieID -> Title).

```python
movies = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/movies.txt")
movies_rdd = movies.map(lambda x: x.split(",")).map(lambda x: ((x[0]), x[1]))
movies_rdd.collect()
```

#### Đọc file ratings_1.txt và ratings_2.txt, map MovieID -> (Rating, 1).

```python
ratings1 = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/ratings_1.txt")
ratings2 = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/ratings_2.txt")
ratings = ratings1.union(ratings2)
ratings.collect()
```

```python
# Map dữ liệu ratings
ratings_rdd = ratings.map(lambda x: x.split(",")).map(lambda x: ((x[1]), (float(x[2]), 1)))
ratings_rdd.collect()
```

#### Reduce để tính tổng điểm và số lượt đánh giá

```python
reduce_ratings = ratings_rdd.reduceByKey(lambda count1, count2: (count1[0] + count2[0], count1[1] + count2[1]))
reduce_ratings.collect()
```

#### Tính điểm trung bình, lọc ra phim có ít nhất 5 lược đánh giá

```python
avg_rating = reduce_ratings.mapValues(lambda x: (x[0] / x[1], x[1]))
avg_rating.collect()
```

```python
# Join với movies để lấy thông tin tên phim
movie_rating = movies_rdd.join(avg_rating)
movie_rating = movie_rating.map(lambda x: (x[1][0], x[1][1][0], x[1][1][1]))
movie_rating.collect()
```

```python
# Lọc phim có ít nhất 5 lượt đánh giá
filter_movie = movie_rating.filter(lambda x: x[2] >= 5)
filter_movie.collect()
```

#### Tìm phim có điểm trung bình cao nhất

```python
# Sắp xếp movie có điểm trung bình từ cao đến thấp
top_movie = filter_movie.sortBy(lambda x: x[1], ascending = False)
for movie in top_movie.collect():
    print(movie)
print("Top movie: ")
top_movie = top_movie.first()
for movie in top_movie:
    print(movie)
```
