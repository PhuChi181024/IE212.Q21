```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
ratings1 = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/ratings_1.txt")
ratings2 = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/ratings_2.txt")
ratings = ratings1.union(ratings2)
ratings.collect()
```

```python
from datetime import datetime
```

```python
rating_year = ratings.map(lambda x: x.split(",")).map(lambda x: (datetime.fromtimestamp(int(x[3])).year, (float(x[2]), 1)))
rating_year.collect()
```

```python
sum_count_rdd = rating_year.reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))
sum_count_rdd.take(5)
```

```python
avg_rdd = sum_count_rdd.mapValues(lambda x: (x[0] / x[1], x[1]))
avg_rdd = avg_rdd.sortByKey()
for row in avg_rdd.collect():
    print(row)
```
