```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

####  Tạo map (UserID -> Gender).

```python
users = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/users.txt")
users_rdd = users.map(lambda x: x.split(",")).map(lambda x: ((x[0]), x[1]))
# users_rdd.collect()
```

```python
ratings1 = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/ratings_1.txt")
ratings2 = sc.textFile("file:///C:/Users/Dell/IE212/Lab3/ratings_2.txt")
ratings = ratings1.union(ratings2)
ratings_rdd = ratings.map(lambda x: x.split(",")).map(lambda x: (x[0], (x[1], float(x[2])))) 
ratings_rdd.collect()
```

#### Join với ratings để thêm thông tin giới tính.

```python
user_rating = ratings_rdd.join(users_rdd)
user_rating.collect()
```

```python
movie_gender_rating = user_rating.map(lambda x: ((x[1][0][0], x[1][1]), x[1][0][1]))
movie_gender_rating.collect()
```

```python
# Tính tổng điểm đánh giá cho mỗi phim theo từng giới tính
sum_rdd = movie_gender_rating.reduceByKey(lambda a, b: a + b)
sum_rdd.collect()
```

```python
# Tính tổng số lượt đánh giá cho mỗi phim theo từng giới tính
count_rdd = movie_gender_rating.mapValues(lambda x: 1).reduceByKey(lambda a, b: a + b)
count_rdd.collect()
```

```python
average_rdd = sum_rdd.join(count_rdd).mapValues(lambda x: (x[0] / x[1]))
average_rdd = average_rdd.sortByKey()
for row in average_rdd.collect():
    print(row)
```
